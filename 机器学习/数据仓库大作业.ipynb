```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import PIL.Image as Image
%matplotlib inline
img = Image.open(r'D:\Py_Project\Langcahin\机器学习\image\img.png')
plt.axis('off')
plt.imshow(img)
```

```python
print(img.info)
row,col = img.size
print("图片大小：",row,col)
```

```python
print('数据类型',type(img))
print('图像的颜色模式：',img.mode)
```

```python
imgData = np.array(img.getdata())
type(imgData)
pixel_vals = imgData.reshape(-1,3)
pixel_vals
km_cluster = KMeans(n_clusters=3)
lable = km_cluster.fit_predict(pixel_vals)
print("每个像素的簇标号：\n",lable)
```

```python
lable = lable.reshape([row,col]).T
pic_new = Image.new("L",(row,col))
for i in range(row):
    for j in range(col):
        pic_new.putpixel((i,j),int(int(256/(lable[i][j]+1))))

plt.imshow(pic_new)
```
